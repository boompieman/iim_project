```python
import xml.etree.ElementTree as ET
```

```python
fileRoot = 'ES2002a.A.words'
```

```python
fileName = '../AMI_corpus_resources/words/' + fileRoot + '.xml'
```

```python
tree = ET.parse(fileName)
```

```python
root = tree.getroot()
```

```python
f = open(fileRoot + '.txt', 'w+')
```

```python
for w in root.findall('w'):
    if str(w.text) == '.' or str(w.text) == ',':
        f.write(str(w.text))
    else:
        f.write(' ' + str(w.text))
```

```python
f.close()
```
